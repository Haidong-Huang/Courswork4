This iterative method is an extension of Newton’s method (the midpoint Newton method) which also starts from an initial estimate $x_0$ and repeatedly gives a new estimate $x_{n+1}$ from the current estimate $x_n$.

The key difference is that this method uses an intermediate evaluation of the derivative to increase the convergence order without requiring second derivatives.
The formula for moving from estimate $x_n$ to $x_{n+1}$ is
$$
x_{n+1}=x_n-\frac{f\left(x_n\right)}{f^{\prime}\left(x_n-\frac{f\left(x_n\right)}{2 f^{\prime}\left(x_n\right)}\right)} \quad \quad \text{(Equation 1)}
$$

Note that you do **not** need to derive this formula. Complete the definition of the function $\mathtt{iterative\_solve}$ so that it returns a 3‑tuple $\left(x_m, f(x_m), m\right)$ where $m$ is the smallest positive integer less than $\mathtt{itermax}$ for which $\left| f(x_m) \right| \leq \text{tol}$ if such an integer exists and $m = \mathtt{itermax}$ otherwise.

Let $p(x)$ be the polynomial $p(x)=x^3-x-1$.
By calling the function $\mathtt{iterative\_solve}$ with appropriate arguments, define a variable $\mathtt{GR}$ that has the value of that approximation to the solution of $p(x)=0$ obtained after performing $m$ iterations of the iterative method starting from the initial estimate $\mathbf{x_0} = 1.5$.

```python
import numpy as np # Do not modify this line.
def iterative_solve(f, df, x_0, tol, itermax ): # Do not modify this line.
    # Start from the given initial estimate.
    x = x_0
    for m in range(1, itermax + 1):
        # Use the formula in the question to get the next estimate.
        fx = f(x)
        midpoint = x - fx / (2 * df(x))
        x = x - fx / df(midpoint)
        # Stop early if the function value is small enough.
        fx = f(x)
        if abs(fx) <= tol or m == itermax:
            return (x, fx, m)

def p(x):
    return x**3 - x - 1

def dp(x):
    return 3*x**2 - 1

# The question asks for the approximation starting from x_0 = 1.5.
GR = iterative_solve(p, dp, 1.5, 1e-12, 100)[0]
```

Complete the definition of the function $\mathtt{modified\_mid}$ so that it returns the numerical value $I$ as given in Equation 2.
Let $g(x)=x \sin (x)$. By calling the function $\mathtt{modified\_mid}$ with appropriate arguments, define a variable $\mathtt{MID}$ that has the value of that approximation to $\int_0^\pi g(x) \mathrm{d} x$ obtained using the modified midpoint rule with $N=30$ subintervals.


```python
import numpy as np  # Do not modify this line.
def modified_mid(f,a,b,N=30):  # Do not modify this line.
    # Split the interval into N equal parts.
    dx = (b - a) / N
    total = 0.0
    for i in range(1, N // 3 + 1):
        # Add the two function values used in each group of three subintervals.
        total += f(a + (3*i - 2) * dx) + f(a + (3*i - 1) * dx)
    # Multiply the sum by the factor in the modified midpoint rule.
    return (3 / 2) * dx * total

def g(x):
    return x * np.sin(x)

# The question asks for the approximation on [0, pi] using N = 30.
MID = modified_mid(g, 0, np.pi, 30)
```
